```python
# To build a URL to a specific function dynamically, use the url_for() function. It accepts the name of the function as
#  its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.
```

```python
from flask import *
myapp= Flask(__name__)


@myapp.route("/")
def test():
    return "Hi welcome to result portal"

@myapp.route("/passed/<int:score>")
def passed(score):
    return "The person has paseed and the marks is " + str(score)

@myapp.route("/fail/<int:score>")
def fail(score):
    return "The person has failed and the marks is " + str(score)


# function containing url_for()

@myapp.route("/check_result/<int:marks>")
def check_result(marks):
    if marks>50:
        return redirect(url_for("passed",score=marks))
    else:
        return redirect(url_for("fail",score=marks))
    

if __name__=="__main__":
    myapp.run(host="0.0.0.0")
```
